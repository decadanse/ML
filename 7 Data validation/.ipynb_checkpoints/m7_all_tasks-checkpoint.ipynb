#### Задание 7.4.13

Допустим, у вас есть выборки y_true = [1.23, 2.35, 2.75] и y_pred = [1.01, 12.3, 2.74]. Посчитайте метрику RMSE, округлите до сотых и запишите результат. Например, 3.14

```python
def print_regression_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f'MSE = {mse:.2f}, RMSE = {rmse:.2f}')
```

```python
y_true = [1.23, 2.35, 2.75]
y_pred = [1.01, 12.3, 2.74]
print_regression_metrics(y_true, y_pred)
```

#### Задание 7.4.17

На этапе вычисления метрик вы получили результаты precision = 0.75 и recall = 0.6. Посчитайте значение метрики f1. Округлите до сотых и запишите результат.

```python
precision = 0.75
recall = 0.6

F1 = 2 * (precision * recall) / (precision + recall)
print(f'f1_score = {F1:.2f}')
```

Целевой переменной будет compliance. Разделите данные на обучающую и тестовую выборки в соотношении 70%/30% без перемешивания. Обучите DecisionTreeClassifier из scikit-learn с параметром random_state=23. Посчитайте значения метрики F1 на тренировочной выборке и на тестовой выборке. Вычтите из значения метрики на тренировочной выборке значение метрики на тестовой выборке и запишите результат в переменную result. 


```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
vis_data = pd.read_csv("./m7.7_train.csv", encoding = 'ISO-8859-1', low_memory = False)
# Напишите ваш код ниже
#vis_data.head()
```

```python
# X = df.drop("Bad Indicator",axis=1)   #Feature Matrix
# Y = df["Bad Indicator"]          #Target Variable
```

```python
vis_data_small = vis_data[['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due', 'compliance']].dropna()
X = vis_data_small[['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due']]
y = vis_data_small['compliance'] #в y всегда содержится целевая переменная
```

```python
# Разбить данные на train/valid
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=False)

# Создать и обучить линейную регрессию    
tree = DecisionTreeClassifier(random_state=23).fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_train)
f1_train = f1_score(y_train, y_pred)

y_pred = tree.predict(X_valid)
f1_test = f1_score(y_valid, y_pred)

result = f1_train - f1_test
```

whats above is correct

```python
print('[train] F1-score = {:.2f}'.format(f1_train))
print('[valid] F1-score = {:.2f}'.format(f1_test))
print('[f1_train - f1_test] result = {:.2f}'.format(result))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.metrics import f1_score, accuracy_score

vis_data = pd.read_csv("./m7.9_train.csv", encoding = 'ISO-8859-1', low_memory = False)
# Напишите ваш код ниже
#vis_data.head()
```

```python
vis_data_small = vis_data[['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due', 'compliance']].dropna()
X = vis_data_small[['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due']]
y = vis_data_small['compliance'] #в y всегда содержится целевая переменная
```

```python
# Разбить данные на train/valid
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=False)

# Создать и обучить линейную регрессию    
tree = DecisionTreeClassifier(random_state=23).fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_train)
f1_train = f1_score(y_train, y_pred)

y_pred = tree.predict(X_valid)
f1_test = f1_score(y_valid, y_pred)
f1_imbalanced = f1_test
```
