```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate

w1 = 10    
a1 = 1     
Fs = 30 
Ts = 1.0/Fs
alpha = 5

##   ( )
t = np.linspace(0, 1, Fs)
##   
y = a1*np.exp(-alpha*t) * np.sin(2*np.pi*w1*t) 
 

##   ( )
tt = np.linspace(0, 1, 100*Fs)
##    (   )
y_high_res = a1*np.exp(-alpha*tt) * np.sin(2*np.pi*w1*tt) 

f, ax = plt.subplots(figsize=(12, 8))
ax.plot(t, y)
ax.plot(tt, y_high_res, '--')
ax.legend(['  ', '  '])
ax.grid()
plt.show()
```

 ####  8.8.1

  "y"   "tt"  .    interpolate.interp1d.           y_high_res   \(MSE=\frac{1}{N}\sum_{1}^{N}(y_i\hat{y}_i)^2\),   5 ,  .

\(MSE=\frac{1}{N}\sum_{1}^{N}(y_i\hat{y}_i)^2\)

```python
from sklearn.metrics import mean_squared_error as mse
f_interp = interpolate.interp1d(t, y, kind='cubic')
np.round(mse(y_high_res, f_interp(tt)), 5)
```

```python
#    .
w1 = 5   
a1 = 1     
Fs = 1000
Ts = 1.0/Fs
alpha = 5

df = pd.DataFrame()
## 
df['t'] = np.linspace(-1, 1, Fs)
##   
df['y'] = np.sin(2*np.pi*w1*df.t) / df.t 
 
##      
df['y_lost'] = df.y
df.loc[0::3, 'y_lost'] = np.nan


f, ax = plt.subplots(figsize=(12, 8))
ax.plot(df.t, df.y_lost)
ax.set_ylim([-10, 10])
ax.legend(['  '])
ax.grid()
plt.show()

df.head(10)
```

####  8.8.2

    y_lost,   interpolate   nearest   Pandas.          "y" \(MSE=\frac{1}{N}\sum_{1}^{N}(y_i\hat{y}_i)^2\),   3   .

```python
df['y_nearest'] = df.y_lost.interpolate(method='nearest')
#   
df.y_nearest.dropna(axis=0)
df.y.dropna(axis=0)
df.y_nearest[np.isnan(df.y_nearest)] = 0 # NaN  0

np.round(mse(df.y, df.y_nearest), 3)
```

####   8.8.3

    y_lost,   interpolate   polynomial, order = 5   Pandas.          "y" \(MSE=\frac{1}{N}\sum_{1}^{N}(y_i\hat{y}_i)^2\),   5 ,  .

```python
df['y_nearest'] = df.y_lost.interpolate(method='polynomial', order = 5)
#   
df.y_nearest.dropna(axis=0)
df.y.dropna(axis=0)

df.y_nearest[np.isnan(df.y_nearest)] = 0 # NaN  0

np.round(mse(df.y, df.y_nearest), 5)
```
