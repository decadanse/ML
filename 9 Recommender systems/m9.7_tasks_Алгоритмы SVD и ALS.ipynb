    user_movie_matrix = np.array(((1,5,0,5,4), (5,4,4,3,2), (0,4,0,0,5), (4,4,1,4,0), (0,4,3,5,0), (2,4,3,5,3))).
    
    2. Сингулярное разложение матрицы — это некое новое представление исходной матрицы, которое отображено в сингулярных числах и сингулярных векторах. Так выглядит формула сингулярного разложения:

        M = U*sigma*V

    где  — исходная матрица «пользователь-фильм»;  — матрица, на диагонали которой лежат сингулярные числа, а вне диагонали — нули;  — матрица-представление пользователей;  — матрица-представление фильмов. 

    Для нас важно понимать, что в лице матриц  и  мы получаем новое представление пользователей и фильмов соответственно. Мы не будем заставлять вас с нуля рассчитывать сингулярные числа и сингулярные векторы, воспользуемся встроенной функцией библиотеки scipy:

    U, s, V = scipy.linalg.svd(M). 

    Задание: При помощи библиотеки scipy проведите SVD-разложение матрицы user_movie_matrix.
    
    3. На этом этапе мы получили новое матричное представление пользователей и фильмов. Заметим, что размерность нашего разложения совпадает с размерностью исходной матрицы. На практике обычно используют сокращённое представление, так как SVD-алгоритм обладает свойством уменьшения размерности без значительной потери информации. На больших матрицах 1 000 000 x 1 000 000 это особенно ощутимо. Для того чтобы сократить размерность, например, до двух компонент, необходимо взять по две первых строки из каждой матрицы-представления.

    Задание: Извлеките из матричных представлений первые две строки и сохраните их. Все дальнейшие действия будут производиться с ними.
    
    4. SVD-разложение на две компоненты произведено. Теперь посмотрим, какой фильм наше новое представление порекомендует в первую очередь для существующего пользователя. 

    Задание: Из исходной матрицы выберем вторую строчку (пользователь №2) —  user_2. 
    
    5. Чтобы перевести пользователя в новое представление сниженной размерности, необходимо его исходный вектор умножить на транспонированную матрицу-представление фильмов
    
            lowdim = user_2*transpose(V)
    
    Задание: Произвести вычисление представления пользователя сниженной размерности.
    
    6. Финальный шаг — необходимо произвести обратную трансформацию вектора в вектор оценок фильмов (т.е. в исходное представление фильмов). Для этого необходимо представление пользователя сниженной размерности умножить на матрицу-представление фильмов
    
            inversed_transformation = lowdim*V

    Задание: Произвести обратную трансформацию вектора сниженной размерности в исходное пространство оценок.
    
    7. Теперь мы получили новый вектор оценок для пользователя №2 (обратите внимание на незначительное изменение оценок фильмов, которые этот пользователь действительно посмотрел). 

```python
import numpy as np
import scipy.linalg
```

```python
#Создадим матрицу «пользователь-фильм», на которой будем практиковаться с SVD-разложением 
user_movie_matrix = np.array(((1,5,0,5,4), (5,4,4,3,2), (0,4,0,0,5), (4,4,1,4,0), (0,4,3,5,0), (2,4,3,5,3)))

#При помощи библиотеки scipy проведите SVD-разложение матрицы user_movie_matrix.
U, s, V = scipy.linalg.svd(user_movie_matrix)

#Извлеките из матричных представлений первые две строки и сохраните их. Все дальнейшие действия будут производиться с ними.
low_U = U[:2,:]
low_V = V[:2,:] 

#Из исходной матрицы выберем вторую строчку (пользователь №2) —  user_2. 
user_2 = user_movie_matrix[2] #(0,4,0,0,5) so 

#Произвести вычисление представления пользователя сниженной размерности.
lowdim = user_2 @ low_V.T #V.T = V.transpose()

#Произвести обратную трансформацию вектора сниженной размерности в исходное пространство оценок
inversed_transformation = lowdim @ low_V
second_user = inversed_transformation
```
